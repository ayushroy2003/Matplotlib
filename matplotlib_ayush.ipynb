Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

Ans - Used to display values for typically two variables for a set of data 


```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(3)  
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))


plt.scatter(x, y)
plt.title("Scatter Plot of Generated Data")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:


The subplot() function in Matplotlib is used to create multiple plots within a single figure. This function allows you to arrange plots in a grid layout, making it easier to compare multiple visualizations side by side.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure
plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title("Line Plot 1")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title("Line Plot 2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title("Line Plot 3")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title("Line Plot 4")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

A bar plot is type of data visualization that represent categorical data using recatangular bars. Each bar length is proportional to the value it represents. Bar plots are used to compare different catrgotical or to show changes over time

It is used because 
- makes it easy to compare across different catgories
- provides a clear and straightforward way to presents data
- Ideal for representing and visualizing cateorical data
- Useful for showing trends over time when categories are ordered 

```python
plt.bar(company, profit)
plt.xlabel("COMPANY")
plt.ylabel("PROFIT")
plt.title("Bar plot")
plt.show()
```

```python
plt.barh(company, profit)
plt.xlabel("COMPANY")
plt.ylabel("PROFIT")
plt.title("horizontal Bar plot")
plt.show()
```
